```python
import torch, os
from PIL import Image
ranking_result_path = '/home/lzj/code/sgg/outputs/ranking_results.pkl'
image_folder = '/home/lzj/datasets/VisualGenome/VG_100K'
relevance_thr = 0.05
ranking_results = torch.load(ranking_result_path)
print(ranking_results.keys())
```

```python
checking_id = 2412290
source_image_path = os.path.join(image_folder, str(checking_id) + '.jpg')
items = sorted(ranking_results[str(checking_id)].items(), key = lambda x: x[1], reverse = True)
items = [key for key, val in items if val > relevance_thr]
target_image_paths = list(map(lambda x: os.path.join(image_folder, x + '.jpg'), items))
if len(target_image_paths) > 10:
    target_image_paths = target_image_paths[:10]
display(Image.open(source_image_path))
for item in target_image_paths:
    display(Image.open(item))
```
